### Q5. 행렬곱 shape 맞추기
1. (2,3) 행렬 A, (3,4) 행렬 B를 만들고 행렬곱을 계산하세요. 결과 shape는?
2. (2,3) 행렬 X와 (2,3) 행렬 Y를 곱하려면 어떻게 해야 하는가?
3. `(2,3) @ (2,3)` 가 왜 에러인지 한 줄로.

행렬곱: `np.dot(A, B)` 또는 `A @ B`  
shape를 맞추려면 한 쪽을 `transpose` 하거나 `reshape` 하면 됨.
</details>

```python
# TODO
# 1
A = np.ones((2,3))
print(A)
B = np.zeros((3,4))
print(B)
dot = np.dot(A,B)
print(dot.shape, '\n')
# 2 transpose 하거나 reshape
# ValueError 행렬곱의 사이즈가 안맞아서
```

```python
A = np.arange(1, 7).reshape(2, 3)         # (2,3)
B = np.arange(1, 13).reshape(3, 4)        # (3,4)
print('A.B shape :', np.dot(A, B).shape)  # (2,4)

X = np.array([[1,2,3],[1,2,3]])
Y = np.array([[1,4,3],[2,4,4]])
print('X . Y.T :\n', np.dot(X, Y.T))      # (2,2)
# (2,3)@(2,3) 은 X의 열(3) != Y의 행(2) 이라 에러
```

### Q6. 선형대수
A = [[1,2],[3,4]], b = [5,6] 일 때

1. 역행렬, 행렬식, 고유값/고유벡터, 랭크, 노름을 모두 출력
2. Ax = b 를 만족하는 x

<details><summary>힌트 (메소드)</summary>

`np.linalg.inv`, `np.linalg.det`, `np.linalg.eig`, `np.linalg.matrix_rank`, `np.linalg.norm`, `np.linalg.solve`
</details>

```python
A = np.array([[1,2],[3,4]])
b = np.array([5, 6])

# TODO
# 시험범위 아님
inv = np.linalg.inv(A)
print(inv)
det = np.linalg.det(A)
print(det)
eig = np.linalg.eig(A)
print(eig)
rank = np.linalg.matrix_rank(A)
print(rank)
norm = np.linalg.norm(A)
print(norm)
print(np.linalg.solve(A,b))
```

```python
print('inv :\n', np.linalg.inv(A))
print('det :', np.linalg.det(A))
vals, vecs = np.linalg.eig(A)
print('eig vals :', vals)
print('eig vecs :\n', vecs)
print('rank :', np.linalg.matrix_rank(A))
print('norm :', np.linalg.norm(A))
print('solve(Ax=b) :', np.linalg.solve(A, b))
```

### Q7. matplotlib - sin & cos
x를 0 ~ 2π 범위에서 만들고, sin은 실선, cos는 점선으로 한 그림에 그리세요. 범례, 제목, x/y 레이블도 모두 표시.

<details><summary>힌트 (메소드)</summary>

`np.arange(0, 2*np.pi, 0.01)`  
`plt.plot(x, y, label=...)`, `linestyle='--'`  
`plt.legend()`, `plt.title()`, `plt.xlabel()`, `plt.ylabel()`, `plt.show()`
</details>

```python
import matplotlib.pyplot as plt
# TODO

x = np.arange(0,2*np.pi, 0.01)
s = np.sin(x)
c = np.cos(x)
plt.plot(x, s, label = 'sin')
plt.plot(x, c, label = 'cos', linestyle = '--')
plt.xlabel('x') ; plt.ylabel('y')
plt.legend(); plt.title("sin & cos")
plt.show()
```
