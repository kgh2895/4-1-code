```python
## CHAPTER 2 퍼셉트론
import numpy as np
x1 = 1
x2 = 2
def perceptron(x1,x2,z):
    z = 100
    w1 = 10 
    w2 = 20
    if z > w1*x1 + w2*x2:
        return 0
    else: return 1

print(perceptron(9,0,100))
print(perceptron(9,1,100))
```

```python
#2.2 AND, NAND, OR, XXXXXXXXXXXXXXXXXXXXXXXXXXXXX

def and_gate(x1,x2):
    if x1*x2 == 1:return 1
    else: return 0

print(and_gate(1,2))
print(and_gate(1,1))
print(and_gate(1,0))
print()
def nand_gate(x1,x2):
    if x1*x2 == 1: return 0
    else: return 1
print(nand_gate(1,1))
print(nand_gate(1,0))
print(nand_gate(0,0))

def or_gate(x1,x2):
    if x1 or x2 == 1: return 1
    else: return 0

print(or_gate(1,0))
print(or_gate(0,0))
```

```python
#2.3 퍼셉트론 구현하기 p.45
def AND(x1,x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = w1*x1 + w2*x2
    if tmp > theta: return 1
    else: return 0

print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1))
```

```python
# 2.5.2 XOR NAND s1 , OR s2
X = np.array([[0,0],[0,1],[1,0],[1,1]])

def NAND_gate(x1,x2):
    if x1 * x2 == 1: return 0
    else: return 1

def OR_gate(x1,x2):
    if x1 == 0 and x2 == 0: return 0
    else: return 1

def XOR(x1,x2):
    s1 = NAND_gate(x1,x2)
    s2 = OR_gate(x1,x2)
    return AND(s1,s2)


for x in X:
    print(NAND_gate(x[0],x[1]))
print()
for x in X:
    print(OR_gate(x[0],x[1]))
print()
for x in X:
    print(XOR(x[0],x[1]))
```
