```python
A = np.array([[1,2],[3,4]])                                                              
                                        
#역행렬                                                              
np.linalg.inv(A)
# [[-2. 1. ]                                                                           
#  [ 1.5 -0.5]]                                           

# 행렬식 (determinant)
np.linalg.det(A)
# -2.0

# 고유값, 고유벡터
vals, vecs = np.linalg.eig(A)
# vals: [-0.37  5.37]  ← 고유값
# vecs: 고유벡터 (열 방향)

# 행렬 랭크
np.linalg.matrix_rank(A)
# 2

# 노름 (벡터/행렬 크기)
np.linalg.norm(A)
# 5.477...

# 연립방정식 풀기 Ax = b
b = np.array([5, 6])
np.linalg.solve(A, b)
# [-4.   4.5]
```

```python
#2.3 퍼셉트론 구현하기 p.45
def AND(x1,x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = w1*x1 + w2*x2
    if tmp > theta: return 1
    else: return 0

print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1))
```

```python
# 2.5.2 XOR NAND s1 , OR s2
X = np.array([[0,0],[0,1],[1,0],[1,1]])

def NAND_gate(x1,x2):
    if x1 * x2 == 1: return 0
    else: return 1

def OR_gate(x1,x2):
    if x1 == 0 and x2 == 0: return 0
    else: return 1

def XOR(x1,x2):
    s1 = NAND_gate(x1,x2)
    s2 = OR_gate(x1,x2)
    return AND(s1,s2)


for x in X:
    print(NAND_gate(x[0],x[1]))
print()
for x in X:
    print(OR_gate(x[0],x[1]))
print()
for x in X:
    print(XOR(x[0],x[1]))
```
