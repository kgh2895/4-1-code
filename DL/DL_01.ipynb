```python
#1.5.6 원소 접근 
x = np.array([[51,55],[14,19],[0,5]])
print('x :', '\n', x, '\n')

print('x[0]:','\n',x[0],'\n')
print('x[0][0]:', '\n', x[0][0],'\n')

for _ in x:
    print(_)
print()
for _ in x:
    for i in _:
        print(i) 
```

```python
#방법 1: zip 사용 (가장 간결)                                               
for col in zip(*x):                                                        
    for i in col:                                                          
        print(i)
print()
#방법 2: 인덱스 사용
for col in range(len(x[0])):
    for row in x:           
        print(row[col]) 
```

```python
# 1.5.4 NumPy 메소드
#array      배열 생성
#zeros,ones 배열 0,1으로 생성
#arange     특정 범위의 숫자를 포함하는 배열 생성
#linspace   특정 범위에서 균등하게 분포된 숫자를 포함하는 배열 
#reshape    배열 크기 변경
#transpose  배열 전치
#sort       배열 정렬
#sum        배열 합
#mean       요소들의 평균값
#std        요소들 표준편차
#max,min    최댓값,최소값
#where      조건에 따라 배열의 요소 선택
#dot        두 배열 내적
#linalg     선형대수 관련함수 
N = np.array([1,2,3])
print('N :', '\n', N)
N = np.zeros((2,2))
print('np.zeros :', '\n', N)
N = np.ones((2,2))
print('np.ones :', '\n', N)
N = np.arange(1,10,2)
print('np.arange(1,10,2)(시작,끝,간격) :', '\n', N)
N = np.linspace(30,0,6)
print('np.linspace(1,10,4)(시작,끝,개수) :', '\n', N)
N = N.reshape(3,2)
print("N.reshape(3,2)", '\n', N)
N = np.transpose(N)
print("np.transpose(N)", '\n', N)
N = np.sort(N)
print('np.sort(N)', '\n', N)
```

```python
print('N.sum(), np.sum(N)', '\n', np.sum(N))
print('N.mean(), np.mean(N)', '\n', N.mean())
print('N.std(), np.std(N)', '\n', N.std())
print('N.max(), np.max(N)', '\n', N.max())
print('N.min(), np.min(N)', '\n', N.min())
print("N", '\n' , N)
print("np.where(N > 12, 1, -1)", '\n', np.where(N > 12, 1, -1))
print('np.where(N > 12, N , 0)', '\n', np.where(N > 12, N, 0))
M = N/6
print('M = N/6', '\n', M)
M = M.reshape((3,2))
print('M.reshape((3,2))', '\n', M)
print('np.dot(N,M)', '\n', np.dot(N,M))
```

```python
A = np.array([[1,2],[3,4]])                                                              
                                        
#역행렬                                                              
np.linalg.inv(A)
# [[-2. 1. ]                                                                           
#  [ 1.5 -0.5]]                                           

# 행렬식 (determinant)
np.linalg.det(A)
# -2.0

# 고유값, 고유벡터
vals, vecs = np.linalg.eig(A)
# vals: [-0.37  5.37]  ← 고유값
# vecs: 고유벡터 (열 방향)

# 행렬 랭크
np.linalg.matrix_rank(A)
# 2

# 노름 (벡터/행렬 크기)
np.linalg.norm(A)
# 5.477...

# 연립방정식 풀기 Ax = b
b = np.array([5, 6])
np.linalg.solve(A, b)
# [-4.   4.5]
```

```python
#pyplot
x = np.arange(0,6,0.01)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y,label = 'sin')
plt.plot(x,z, label = 'cos')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title('sin&cos')
plt.show()
```

```python
#1.6.2 pyplot 기능 
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,6,0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label = 'sin')
plt.plot(x, y2, linestyle ='--', label = 'cos')
plt.xlabel("x")
plt.ylabel('y')
plt.legend
plt.title("sin & cos")
plt.show()
```
