```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1.0,2.0,3.0])
print(x)
type(x)
```

```python
x = np.array([1.0,2.0,3.0])
y = np.array([2.0,4.0,8])

print(x+y)
print(x-y)
print(x*y)
print(x/y)
```

```python
x = np.array([[1,2],[4,5]])
print(x)
print()
print(x.shape)
print(x.dtype)

print()
y = np.array([[3,0],[6,0]])
print(x+y)
print(x*y)

```

```python
a = np.array([[1,2],[3,4]])
b = np.array([3,4])
c = np.array([[3,4]])

print(a+b)
print(a*b)
print()
print(a+c)
print(a*c)
```

```python
x = np.array([[51,55],[14,19],[0,5]])
print(x)
print()
print(x[0])
print(x[0][0])
print()
for _ in x:
    print(_)
print()
for _ in x:
    for i in _:
        print(i) 
```

```python
#방법 1: zip 사용 (가장 간결)                                               
for col in zip(*x):                                                        
    for i in col:                                                          
        print(i)
print()
#방법 2: 인덱스 사용
for col in range(len(x[0])):
    for row in x:           
        print(row[col]) 
```

```python
x = np.arange(0,6,0.01)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y)
plt.plot(x,z)
plt.show()
```
